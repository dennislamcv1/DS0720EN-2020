```python
#Replace year 0 with 2020
df['yearbuilt'].replace(to_replace=0, value=2020,inplace=True)
```

```python
df['yearbuilt'].value_counts()
```

```python
#Create a new column year
df['year'] = 2020 - df['yearbuilt'].astype('int')
```
