```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(font_scale=1.25)
np.random.seed(5)

pd.set_option("display.max_rows",10000)
pd.set_option("display.max_columns",100)
```

```python
#Load first dataset
df1 = pd.read_csv("nycbronx2.csv")
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.shape
```

```python
df1['borough'].value_counts()
```

### Drop all features except complaint_type, incident_zip, status

```python
df1 = df1[['incident_zip','status','complaint_type']]
```

```python
df1.head()
```

```python
df1.shape
```

### Load the second dataset and sample 17330 samples

```python
df2 = pd.read_csv("bronx.csv")
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2['borough'].value_counts()
```

### Drop unnecessary features

```python
df2 = df2.drop(['borough','lot','zipcode','address','yearalter1','latitude','longitude'],axis=1)
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

### Drop all NaNs

```python
#Discard all NaNs
df2.dropna(inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
df2.shape
```

### Drop duplicates

```python
df2.drop_duplicates(inplace=True)
```

```python
df2.duplicated().sum()
```

```python
df2.shape
```

```python
df2.head()
```

### Create Sampling from population

```python
df2 = df2.sample(n=33759)
```

```python
df2.head()
```

```python
df2.shape
```

### Merge the two datasets

```python
df3 = df2.join(df1,lsuffix='_left',rsuffix='_right',how='left')
```

```python
df3
```

```python
df3.isnull().sum()
```

```python
#Discard all NaNs
df3.dropna(inplace=True)
```

```python
df3.head()
```

```python
df3.isnull().sum()
```

```python
df3.shape
```
